```python
def count_lines(cell):
    if isinstance(cell, str):
        return len(cell.split("\n"))
    return cell

subQlen = df.iloc[:, 1::2].applymap(count_lines)
subQlen = subQlen.applymap(lambda x: x if x > 1 else np.nan)

```

```python
query_type = "pin_FOL_NL"
```

```python
subQlen[query_type].dropna()
```

```python
df[query_type][subQlen[query_type].dropna().index].values
```

```python
# df.loc[:, query_type] = df.loc[:, query_type].apply(lambda x: x.replace("\n\n", ' / ') if isinstance(x, str) else x)
df.loc[:, query_type] = df.loc[:, query_type].apply(lambda x: x.replace("Here is the natural language query:\n\n", '') if isinstance(x, str) else x)
df.loc[:, query_type] = df.loc[:, query_type].apply(lambda x: x.replace("Here is the converted natural language query:\n\n", '') if isinstance(x, str) else x)
# df.loc[:, query_type] = df.loc[:, query_type].apply(lambda x: x.replace("Sub-queries:\n", '') if isinstance(x, str) else x)
df.loc[:, query_type] = df.loc[:, query_type].apply(lambda x: x.replace("\n", '') if isinstance(x, str) else x)
# df.loc[:, query_type] = df.loc[:, query_type].apply(lambda x: x.replace(",.", '?') if isinstance(x, str) else x)
```
