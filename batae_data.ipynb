```python
import pandas as pd
import json
```

```python
id2ent = pd.read_pickle("data/FB15k-237-betae/id2ent.pkl")
id2rel = pd.read_pickle("data/FB15k-237-betae/id2rel.pkl")
with open ("data/ent2txt.json") as f:
    ent2txt = json.load(f)
id2rel
```

```python
query_structure_to_type = {('e', ('r',)): '1p',
                           ('e', ('r', 'r')): '2p',
                           ('e', ('r', 'r', 'r')): '3p',
                           (('e', ('r',)), ('e', ('r',))): '2i',
                           (('e', ('r',)), ('e', ('r',)), ('e', ('r',))): '3i',
                           ((('e', ('r',)), ('e', ('r',))), ('r',)): 'ip',
                           (('e', ('r', 'r')), ('e', ('r',))): 'pi',
                           (('e', ('r',)), ('e', ('r', 'n'))): '2in',
                           (('e', ('r',)), ('e', ('r',)), ('e', ('r', 'n'))): '3in',
                           ((('e', ('r',)), ('e', ('r', 'n'))), ('r',)): 'inp',
                           (('e', ('r', 'r')), ('e', ('r', 'n'))): 'pin',
                           (('e', ('r', 'r', 'n')), ('e', ('r',))): 'pni',
                           (('e', ('r',)), ('e', ('r',)), ('u',)): '2u',
                           ((('e', ('r',)), ('e', ('r',)), ('u',)), ('r',)): 'up',
                           # ((('e', ('r', 'n')), ('e', ('r', 'n'))), ('n',)): '2u-DM',
                           # ((('e', ('r', 'n')), ('e', ('r', 'n'))), ('n', 'r')): 'up-DM'
                           }
```

# 원본 데이터

```python
query = pd.read_pickle("data/FB15k-237-betae/train-queries.pkl")
query_answer = pd.read_pickle("data/FB15k-237-betae/train-answers.pkl")
query
```

```python
query_answer[(3793, (107,))	]
```

```python
query_type = []
queries = []
for j in range(10):
    for i in list(list(query.values())[j]):
        queries.append(i)
        query_type.append(query_structure_to_type[list(query.keys())[j]])

queries
```

```python
answers = []
for q in queries:
    answers.append(query_answer[q])
answers
```

```python
qa = pd.DataFrame({"types": query_type,
    "queries": queries,
    "answers": answers,
                   })

qa
```

```python
# pni
qa[qa["types"] == "pin"]["queries"].apply(lambda x: x if x[1][1][0]+1 in set(x[0][1]) else None).dropna()
```

```python
# pni
qa[qa["types"] == "pin"]["queries"].apply(lambda x: x if ((x[0][1][0] % 2 == 0 and x[0][1][1] > x[0][1][0]) or (x[0][1][0] % 2 != 0 and x[0][1][1] < x[0][1][0])) and abs(x[0][1][0] - x[0][1][1]) == 1 else None).dropna()
```
