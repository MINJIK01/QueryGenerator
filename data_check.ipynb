```python
made = pd.read_csv("data/sample_20k_0007.csv")
org_train = pd.read_csv("data/FB15k-237-train.csv")
org_valid = pd.read_csv("data/FB15k-237-valid.csv")
org_test = pd.read_csv("data/FB15k-237-test.csv")
```

```python
print(made.columns)
print(org_train.columns)
print(org_valid.columns)
print(org_test.columns)
```

```python
made = made.iloc[:, 2::4].reset_index()
org_train = org_train.iloc[:, 20::2].reset_index()
org_valid = org_valid.iloc[:, 28::3].reset_index()
org_test = org_test.iloc[:, 28::3].reset_index()

print(made.columns)
print(org_train.columns)
print(org_valid.columns)
print(org_test.columns)
```

```python
reasoning_data = {
    # "train": org_train,
    # "valid": org_valid,
    "test": org_test,
}

overlapped_x = []
overlapped_y = []
for k, v in reasoning_data.items():

    # 교집합 구하기 (두 데이터프레임에 모두 있는 데이터)
    # for i in range(1, len(v.columns)):
    for i in range(2, 3):
        overlap_data = made.merge(v, on=made.columns[i], how='inner')
        print(len(overlap_data['index_x'].tolist()))
        print(len(overlap_data['index_y'].tolist()))
        overlapped_x.append(overlap_data['index_x'].tolist())
        overlapped_y.append(overlap_data['index_y'].tolist())

print("두 데이터프레임에 모두 있는 데이터:")
print(overlapped_x)
print(len(overlapped_x))

print(overlapped_y)
print(len(overlapped_y))
```
