#### FOL 중복 확인

```python
add = pd.read_csv("data/sample_add.csv")
made = pd.read_csv("data/sample_20k_0007.csv")

org_train = pd.read_csv("data/FB15k-237-train.csv")
org_valid = pd.read_csv("data/FB15k-237-valid.csv")
org_test = pd.read_csv("data/FB15k-237-test.csv")
```

```python
print(made.columns)
print(org_train.columns)
print(org_valid.columns)
print(org_test.columns)
```

```python
added = add.iloc[:, 2::4].reset_index()
maded = made.iloc[:, 2::4].reset_index()
org_trained = org_train.iloc[:, 20::2].reset_index()
org_valided = org_valid.iloc[:, 28::3].reset_index()
org_tested = org_test.iloc[:, 28::3].reset_index()

print(added.columns)
print(maded.columns)
print(org_trained.columns)
print(org_valided.columns)
print(org_tested.columns)
```

```python
reasoning_data = {
    # "made": made,
    "train": org_trained,
    "valid": org_valided,
    "test": org_tested,
}

overlapped_x = []
overlapped_y = []
for k, v in reasoning_data.items():

    # 교집합 구하기 (두 데이터프레임에 모두 있는 데이터)
    for i in range(2, len(v.columns)):
        overlap_data = maded.merge(v, on=v.columns[i], how='inner')
        print(len(overlap_data['index_x'].tolist()))
        print(len(overlap_data['index_y'].tolist()))
        try:
            made.drop(overlap_data['index_x'].tolist(), inplace=True)
        except:
            pass
        overlapped_x.append(overlap_data['index_x'].tolist())
        overlapped_y.append(overlap_data['index_y'].tolist())

print("두 데이터프레임에 모두 있는 데이터:")
print(overlapped_x)
print(len(overlapped_x))

print(overlapped_y)
print(len(overlapped_y))
```

```python
# 교집합 구하기 (두 데이터프레임에 모두 있는 데이터)
overlap_data = made_nl.merge(nls, on="query", how='inner')

print("두 데이터프레임에 모두 있는 데이터:")
print(overlap_data['id_x'].tolist())
overlap_data
```
